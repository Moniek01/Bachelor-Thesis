```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
FallAllD = pd.read_pickle('FallAllD.pkl')
FallAllD['Fall'] = (FallAllD['ActivityID'] > 100).astype(int)
print('Rows:', len(FallAllD))
FallAllD.head()
```

```python
SubjectID_amount = FallAllD['SubjectID'].value_counts().sort_index()
    
plt.figure()
SubjectID_amount.plot(kind='bar')
```

```python
Device_amount = FallAllD['Device'].value_counts()
print(Device_amount)

plt.figure()
Device_amount.plot(kind='bar')
```

```python
ActivityID_amount = FallAllD['ActivityID'].value_counts().sort_index()
    
plt.figure(figsize=(16,5))
ActivityID_amount.plot(kind='bar')
```

```python
TrialNo_amount = FallAllD['TrialNo'].value_counts().sort_index()
    
plt.figure()
TrialNo_amount.plot(kind='bar')
```

```python
Fall_amount = FallAllD['Fall'].value_counts().sort_index()
print(Fall_amount)
print(round(Fall_amount[1] / len(FallAllD) * 100, 2), 'percent fall activities')
    
plt.figure()
Fall_amount.plot(kind='bar')
```

```python
Neck = FallAllD[FallAllD['Device'] == 'Neck']
Fall_amount = Neck['Fall'].value_counts().sort_index()
print(Fall_amount)
print(round(Fall_amount[1] / len(Neck) * 100, 2), 'percent fall activities - neck')

Waist = FallAllD[FallAllD['Device'] == 'Waist']
Fall_amount = Waist['Fall'].value_counts().sort_index()
print(Fall_amount)
print(round(Fall_amount[1] / len(Waist) * 100, 2), 'percent fall activities - waist')

Wrist = FallAllD[FallAllD['Device'] == 'Wrist']
Fall_amount = Wrist['Fall'].value_counts().sort_index()
print(Fall_amount)
print(round(Fall_amount[1] / len(Wrist) * 100, 2), 'percent fall activities - wrist')
```
