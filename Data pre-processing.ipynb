```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
FallAllD = pd.read_pickle('FallAllD.pkl')
FallAllD = FallAllD[FallAllD['ActivityID'] > 0]
FallAllD['Fall'] = (FallAllD['ActivityID'] > 100).astype(int)
FallAllD.drop('Mag', axis=1, inplace=True)
FallAllD.drop('Bar', axis=1, inplace=True)
FallAllD.drop('ActivityID', axis=1, inplace=True)
FallAllD.drop('TrialNo', axis=1, inplace=True)
FallAllD.head()
```

```python
# Accelerator mean
FallAllD['Acc_x_mean'] = FallAllD['Acc'].apply(lambda x: np.mean(x[:, 0]))
FallAllD['Acc_y_mean'] = FallAllD['Acc'].apply(lambda y: np.mean(y[:, 1]))
FallAllD['Acc_z_mean'] = FallAllD['Acc'].apply(lambda z: np.mean(z[:, 2]))
# Accelerator standard deviation
FallAllD['Acc_x_sd'] = FallAllD['Acc'].apply(lambda x: np.std(x[:, 0]))
FallAllD['Acc_y_sd'] = FallAllD['Acc'].apply(lambda y: np.std(y[:, 1]))
FallAllD['Acc_z_sd'] = FallAllD['Acc'].apply(lambda z: np.std(z[:, 2]))
# Accelerator range
FallAllD['Acc_x_range'] = FallAllD['Acc'].apply(lambda x: np.max(x[:, 0]) - np.min(x[:, 0]))
FallAllD['Acc_y_range'] = FallAllD['Acc'].apply(lambda y: np.max(y[:, 1]) - np.min(y[:, 1]))
FallAllD['Acc_z_range'] = FallAllD['Acc'].apply(lambda z: np.max(z[:, 2]) - np.min(z[:, 2]))

FallAllD.drop('Acc', axis=1, inplace=True)
```

```python
# Gyroscope mean
FallAllD['Gyr_x_mean'] = FallAllD['Gyr'].apply(lambda x: np.mean(x[:, 0]))
FallAllD['Gyr_y_mean'] = FallAllD['Gyr'].apply(lambda y: np.mean(y[:, 1]))
FallAllD['Gyr_z_mean'] = FallAllD['Gyr'].apply(lambda z: np.mean(z[:, 2]))
# Gyroscope standard deviation
FallAllD['Gyr_x_sd'] = FallAllD['Gyr'].apply(lambda x: np.std(x[:, 0]))
FallAllD['Gyr_y_sd'] = FallAllD['Gyr'].apply(lambda y: np.std(y[:, 1]))
FallAllD['Gyr_z_sd'] = FallAllD['Gyr'].apply(lambda z: np.std(z[:, 2]))
# Gyroscope range
FallAllD['Gyr_x_range'] = FallAllD['Gyr'].apply(lambda x: np.max(x[:, 0]) - np.min(x[:, 0]))
FallAllD['Gyr_y_range'] = FallAllD['Gyr'].apply(lambda y: np.max(y[:, 1]) - np.min(y[:, 1]))
FallAllD['Gyr_z_range'] = FallAllD['Gyr'].apply(lambda z: np.max(z[:, 2]) - np.min(z[:, 2]))

FallAllD.drop('Gyr', axis=1, inplace=True)

FallAllD.head()
```

```python
Neck = FallAllD[FallAllD['Device'] == 'Neck'].copy()
Neck.drop('Device', axis=1, inplace=True)

Neck.head()
```

```python
grouped_df = Neck.groupby(['SubjectID', 'Fall']).size().unstack(fill_value=0)

# Plotting
ax = grouped_df.plot(kind='bar', stacked=False, figsize=(10, 6))
ax.set_ylabel('Frequency')
ax.set_xlabel('Subject ID')
ax.set_title('Fall Counts by Subject ID - Neck')
plt.legend(['No Fall', 'Fall'])
plt.show()
```

```python
# Subjects 4 and 10 don't have any fall activities
Neck = Neck[Neck['SubjectID'] != 4]
Neck = Neck[Neck['SubjectID'] != 10]
Neck = Neck.reset_index(drop=True)

grouped_df = Neck.groupby(['SubjectID', 'Fall']).size().unstack(fill_value=0)
# Plotting
ax = grouped_df.plot(kind='bar', stacked=False, figsize=(10, 6))
ax.set_ylabel('Frequency')
ax.set_xlabel('Subject ID')
ax.set_title('Fall Counts by Subject ID - Neck')
plt.legend(['No Fall', 'Fall'])
plt.show()
```
